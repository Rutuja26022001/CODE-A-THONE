#### calculate_median

```python
numbers = [ 4,10,29,33,42,67,44, ]
def calculate_median(list_of_number):
    n = len(list_of_number)
    median = n//2
    if n% 2:
        return sorted(list_of_number)[median]
    return sum(sorted(list_of_number)[median-1:median-2])/2
        
print("The Median of given numbers is :" ,calculate_median(numbers))      
```

#### calculate_mode

```python
def find_most_frequent_words(A):
    with open(A) as f:
        lines = f.readlines()
    line=str(lines)
    split_list= line.split()
    alpha=[]
    for i in split_list:
        a= i.isalpha()
        if a==True:
            alpha.append(i)
    import pandas as pd
    count = pd.Series(alpha).value_counts().sort_values(ascending = False)
    print('10 most frequent words: ')
    print(count.head(10))
find_most_frequent_words('obama_speech.txt')
    
```

```python
import wordcloud
from wordcloud import WordCloud,STOPWORDS, ImageColorGenerator
from PIL import Image
import numpy as np
filename='rom_jul.png'
with Image.open(filename) as img:
    img.load()
img.show()
img_gray=img.convert('L')
threshold=125
img_threshold=img_gray.point(
    lambda x:255 if x>threshold else 0
)
mask_img=np.array(img_threshold)
comment_words=''
stopwords=set(STOPWORDS)
text=open('romeo_and_juliet.txt').read()
Text=str(text)
wc=WordCloud(width=3000,
                     height=3000,
                     stopwords=stopwords, 
                     max_words=200,
                     mask=mask_img,
                     min_font_size=10).generate(text)
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(16,20))
plt.imshow(wc,
           interpolation='bilinear')
plt.axis('off')
plt.show()
```
